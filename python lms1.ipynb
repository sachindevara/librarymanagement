```python
class Book:
    def __init__(self, title, author, quantity):
        self.title = title
        self.author = author
        self.quantity = quantity

class Library:
    def __init__(self):
        self.books = []

    def add_book(self):
        title = input("Enter book title: ")
        author = input("Enter book author: ")
        quantity = int(input("Enter quantity: "))
        book = Book(title, author, quantity)
        self.books.append(book)
        print(f"{quantity} copies of {title} by {author} have been added to the library.")

    def view_books(self):
        if len(self.books) == 0:
            print("There are no books in the library.")
        else:
            print("List of books in the library:")
            for book in self.books:
                print(f"{book.title} by {book.author} - {book.quantity} copies")

    def search_book(self):
        title = input("Enter book title to search: ")
        for book in self.books:
            if book.title == title:
                print(f"{book.title} by {book.author} - {book.quantity} copies are available in the library.")
                return
        print(f"{title} is not available in the library.")

    def remove_book(self):
        title = input("Enter book title to remove: ")
        for book in self.books:
            if book.title == title:
                self.books.remove(book)
                print(f"{book.title} by {book.author} has been removed from the library.")
                return
        print(f"{title} is not available in the library.")

class User:
    def __init__(self, username, password):
        self.username = username
        self.password = password

class Customer(User):
    def request_book(self):
        title = input("Enter book title to request: ")
        for book in lib.books:
            if book.title == title:
                if book.quantity > 0:
                    book.quantity -= 1
                    print(f"{book.title} by {book.author} has been borrowed.")
                    return
                else:
                    print(f"Sorry, {book.title} by {book.author} is currently out of stock.")
                    return
        print(f"{title} is not available in the library.")

    def return_book(self):
        title = input("Enter book title to return: ")
        for book in lib.books:
            if book.title == title:
                book.quantity += 1
                print(f"{book.title} by {book.author} has been returned.")
                return
        print(f"{title} is not a valid book to return.")

class Admin(User):
    def add_book(self):
        lib.add_book()

    def view_books(self):
        lib.view_books()

    def search_book(self):
        lib.search_book()

    def remove_book(self):
        lib.remove_book()

users = {
    "admin": "admin123"
}
lib = Library()

while True:
    print("\nWelcome to the Library Management System")
    print("1. Login")
    print("2. Sign up")
    print("3. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        username = input("Enter username: ")
        password = input("Enter password: ")
        if username in users and users[username] == password:
            if username == "admin":
                admin = Admin(username, password)
                while True:
                    print("\nAdmin Panel")
                    print("1. Add Book")
                    print("2. View Books")
                    print("3. Search Book")
                    print("4. Remove Book")
                    print("5. Logout")
                    choice = int(input("Enter your choice: "))

                    if choice == 1:
                        admin.add_book()
                    elif choice == 2:
                         admin.view_books()
                    elif choice == 3:
                         admin.search_book()
                    elif choice == 4:
                         admin.remove_book()
                    elif choice == 5:
                         break
                    else:
                         print("Invalid choice. Try again.")

            else:
                customer = Customer(username, password)
                while True:
                    print("\nCustomer Panel")
                    print("1. Request Book")
                    print("2. Return Book")
                    print("3.view books")
                    print("4. Logout")

                    choice = int(input("Enter your choice: "))

                    if choice == 1:
                         customer.request_book()
                    elif choice == 2:
                         customer.return_book()
                    elif choice == 3:
                         admin.view_books()
                    elif choice==4:
                         break
                    else:
                         print("Invalid choice. Try again.")

        else:
            print("Invalid username or password. Try again.")

    elif choice == 2:
        username = input("Enter username: ")
        if username in users:
            print("Username already exists. Try a different username.")
            continue
        password = input("Enter password: ")
        users[username] = password
        print("Sign up successful. Please login to continue.")

    elif choice == 3:
        print("Thank you for using the Library Management System.")
        break

    else:
        print("Invalid choice. Try again.")
```
